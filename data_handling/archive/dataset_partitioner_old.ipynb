```python
import os
import shutil
import random
random.seed(42)
```

```python
dataset_path = '/home/martin/Projects/ongoing/ukw/data/base/full/total/'
new_dataset_dir_part = '/home/martin/Projects/ongoing/ukw/data/base/full/total/incipiente'


dataset_labels = [dataset_path + 'train/labels/' + file for file in sorted(os.listdir(dataset_path + 'train/labels/'))] + [dataset_path + 'test/labels/' + file for file in sorted(os.listdir(dataset_path + 'test/labels/'))] + [dataset_path + 'valid/labels/' + file for file in os.listdir(dataset_path + 'valid/labels/')]
dataset_images = [dataset_path + 'train/images/' + file for file in sorted(os.listdir(dataset_path + 'train/images/'))] + [dataset_path + 'test/images/' + file for file in sorted(os.listdir(dataset_path + 'test/images/'))] + [dataset_path + 'valid/images/' + file for file in os.listdir(dataset_path + 'valid/images/')]

indices = list(range(len(dataset_images)))
random.shuffle(indices)

elements = ["images", "labels"]
set_types = ["test", "train", "valid"] 
```

```python
for part_num in range(1,3):
    # Define the root directory
    root_dir = f"{new_dataset_dir_part}_{part_num}"

    # Define subdirectories
    subdirectories = ["test", "train", "valid"]

    # Define sub-subdirectories
    subsubdirectories = ["images", "labels"]

    # Create the root directory if it doesn't exist
    if not os.path.exists(root_dir):
        os.mkdir(root_dir)

    # Create subdirectories and sub-subdirectories
    for subdirectory in subdirectories:
        subdir_path = os.path.join(root_dir, subdirectory)
        if not os.path.exists(subdir_path):
            os.mkdir(subdir_path)
        
        # Create sub-subdirectories
        for subsubdirectory in subsubdirectories:
            subsubdir_path = os.path.join(subdir_path, subsubdirectory)
            if not os.path.exists(subsubdir_path):
                os.mkdir(subsubdir_path)
```

```python
def divide_list_into_two(input_list):
    # Calculate the length of each sublist
    total_length = len(input_list)
    length_sublist = total_length // 2
    print(total_length)
    print(length_sublist)
    
    # Divide the input list into two equal-sized sublists
    sublist1 = input_list[:length_sublist]
    # # Allocate remaining elements to the third sublist
    sublist2 = input_list[length_sublist:]
    
    # sublist3 = input_list[length_sublist * 2:]
    
    return [sublist1, sublist2]#, sublist3]


def divide_list_with_ratios(input_list, indices, ratio1, ratio2, ratio3):
    # Calculate the lengths of the resulting lists based on ratios
    total_length = len(input_list)
    length1 = int(total_length * ratio1)
    length2 = int(total_length * ratio2)
    
    # Divide the shuffled indices into three lists based on the calculated lengths
    list1_indices = indices[:length1]
    print(list1_indices)
    list2_indices = indices[length1:]
    print(list2_indices)
    list3_indices = indices[length1 + length2:]
    print(list3_indices)
    
    # Create lists with corresponding elements from the original list
    list1 = divide_list_into_two([input_list[i] for i in list1_indices])
    list2 = divide_list_into_two([input_list[i] for i in list2_indices])
    list3 = divide_list_into_two([input_list[i] for i in list3_indices])
    
    return list1, list2, list3
```
